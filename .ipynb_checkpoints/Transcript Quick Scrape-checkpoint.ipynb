```python
import urllib
import csv
savefile = urllib.URLopener()
```

```python
def get_vtt(url, title):
    file_out_type = title + ".vtt"
    try:
        savefile.retrieve(url, file_out_type)
        print("file saved")
    except:
        try:
            savefile.retrieve(url, file_out_type)
            print("file saved")
        except:
            print(title) 
```

```python
with open('newshour-vtt.csv', 'rb') as csvfile:
    nhreader = csv.reader(csvfile, dialect = "excel")
    next(nhreader)
    for row in nhreader:
        url = row[0]
        print(url)
        print(title)
        title = row[1]
        get_vtt(url, title)
```
